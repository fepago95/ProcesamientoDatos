### calculas imc vc cal icfes

```python
estudiantes['imc']=estudiantes['peso']
```

```python
estudiantes.shape
```

```python
estudiantes.shape[0]
```

```python
for i in range(0,estudiantes.shape[0]):
  estudiantes['imc'][i]=(estudiantes['peso'][i])/(((estudiantes['estatura'][i])/100)*((estudiantes['estatura'][i])/100))
```

```python
sns.histplot(data=estudiantes['imc'], palette='Pastel1')

plt.xlabel('Índice de masa corporal')
plt.ylabel('Frecuencia')
plt.title('Histograma de índice de masa corporal')
plt.show();
```

```python
fig, axes=plt.subplots(1,1, figsize=(12,6))

sns.scatterplot(x=estudiantes['imc'], y=estudiantes['cal_icfes'], palette='Pastel1', ax=axes)

axes.set_title('IMC vs Calificación ICFES')
axes.set_xlabel('IMC')
axes.set_ylabel('Calificación ICFES')
plt.show();
```
